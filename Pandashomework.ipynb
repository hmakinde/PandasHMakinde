```python
### Gender Demographics

#First remove duplicates based on the total unique users
uniques = gamesdf.drop_duplicates(subset='SN', keep="first")
uniques.head(20)
#check your new df to make sure its changed
describe = uniques.describe()
describe
```

```python
#Then calculate the gender stats

# First Create a group based on the unique values in the 'gender' column
playa_gender = uniques.groupby(by = "Gender")
# then Count how many times each gender appears in the data
count_gender = playa_gender["Gender"].count()
count_gender

Gamegenders = pd.DataFrame(count_gender)
Gamegenders
```

```python
#An easier way to get the percentage of all genders at once
gender_percent = count_gender / 576 * 100
gender_percent = gender_percent.round(2)
gender_percent
```

```python
#join both tables to make a new dataframe
Genderdemo_df = Gamegenders.join(gender_percent, lsuffix='_Count', rsuffix='_Percent')
Genderdemo_df
```

```python
### Purchasing Analysis (Gender)
#First group whole df by gender
all_gender = gamesdf.groupby(by = "Gender")
all_gender.head(3)
```

```python
#Purchase Count calculate all purchases based on gender group duplicates included
Buycount_gender = all_gender["Gender"].count()
Buycount_gender
Gender_purchase_df = pd.DataFrame({"Purchase Count" : Buycount_gender})
```

```python
# Total Purchase Value break down total purchase by gender
BuyRevenue_gender = all_gender['Price'].sum()
BuyRevenue_gender =BuyRevenue_gender.round(2)
BuyRevenue_gender_df = pd.DataFrame({"Total Purchase" :BuyRevenue_gender})
```

```python
# Average Purchase Price  total purchase divided by the counts
Buyavg_gender = all_gender["Price"].mean()
Buyavg_gender = Buyavg_gender.round(2)
Buyavg_gender_df = pd.DataFrame({"Average Purchase Price" : Buyavg_gender})
```

```python
# Average Purchase Total per Person by Gender
averageper_person = BuyRevenue_gender / count_gender
averageper_person = averageper_person.round(2)
averageper_person_df = pd.DataFrame({"Average per Person" : averageper_person})
```

```python
#join all the dataframes together
Genderpurch_df = Gender_purchase_df.join(BuyRevenue_gender_df)
Genderpurchb_df = Buyavg_gender_df.join(averageper_person_df)
GendPurch_table = Genderpurch_df.join(Genderpurchb_df)

#add dollar signs
GendPurch_table["Total Purchase"] = '$' + GendPurch_table["Total Purchase"].astype(str) 
GendPurch_table["Average Purchase Price"] = '$' + GendPurch_table["Average Purchase Price"].astype(str) 
GendPurch_table["Average per Person"] = '$' + GendPurch_table["Average per Person"].astype(str)


#display your table
GendPurch_table
```

```python
### Age Demographics
# Figure out the minimum and maximum ages for the gamers
print(gamesdf["Age"].max())
print(gamesdf["Age"].min())


# Create bins in which to place values based upon TED Talk views
bins = [0, 9, 14, 19, 24, 29, 34, 39, 45]

# Create labels for these bins
group_labels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]


# Slice the data and place it into bins then add the series into a new column inside of the DataFrame
gamesdf["Age-Group"] = pd.cut(gamesdf["Age"], bins, labels=group_labels)
gamesdf.head()
```

```python
#now repeat everything from the purchasing analysis by gender, but do it by age
agegrouping = gamesdf.groupby(by = "Age-Group")
agegrouping.head()

#get unige ages in the group
uniques2 = uniques
uniques2["Age-Group"] = pd.cut(uniques2["Age"], bins, labels=group_labels)
#group them by age
playa_age = uniques2.groupby(by = "Age-Group")
count_age = playa_age["Age-Group"].count()
count_age
```

```python
#Purchase Count calculate all purchases based on gender group duplicates included
Buycount_age= agegrouping["Age-Group"].count()
Buycount_age
Age_purchase_df = pd.DataFrame({"Purchase Count" : Buycount_age})

# Total Purchase Value break down total purchase by age group
BuyRevenue_age = agegrouping["Price"].sum()
BuyRevenue_age =BuyRevenue_age.round(2)
BuyRevenue_age_df = pd.DataFrame({"Total Purchase" : BuyRevenue_age})

# Average Purchase Price  total purchase divided by the counts by age group
Buyavg_age = agegrouping["Price"].mean()
Buyavg_age = Buyavg_age.round(2)
Buyavg_age_df = pd.DataFrame({"Average Purchase Price" : Buyavg_age})

# Average Purchase Total per Person by age group

averageper_age = BuyRevenue_age / count_age
averageper_age = averageper_age.round(2)
averageper_age_df = pd.DataFrame({"Average per Age group" : averageper_age})
```

```python
#join all the dataframes together
Agepurch_df = Age_purchase_df.join(BuyRevenue_age_df)
Agepurchb_df = Buyavg_age_df.join(averageper_age_df)
AgePurch_table = Agepurch_df.join(Agepurchb_df)

AgePurch_table["Total Purchase"] = '$' + AgePurch_table["Total Purchase"].astype(str) 
AgePurch_table["Average Purchase Price"] = '$' + AgePurch_table["Average Purchase Price"].astype(str) 
AgePurch_table["Average per Age group"] = '$' + AgePurch_table["Average per Age group"].astype(str)

AgePurch_table
```

```python
### Top Spenders

 #Identify the the top 5 spenders in the game by total purchase value, then list (in a table):

 # First Create a group based on the unique values in the 'SN users' column 
playas = gamesdf.groupby(by = "SN")
playas.head(5)
 
```

```python
#Purchase Count calculate all purchases based on SN of the players 
count_playas = playas["Purchase ID"].count()
count_playasdf = pd.DataFrame({"Player Counts" : count_playas})
```

```python
#Average Purchase Price 
Buyavg_playas = playas["Price"].mean()
Buyavg_playas = Buyavg_playas.round(2)
Buyavg_playas_df = pd.DataFrame({"Average Purchase Price" : Buyavg_playas})
```

```python

# Total Purchase Value break down total purchase by players
Revenue_playas = playas['Price'].sum()
Revenue_playas = Revenue_playas.round(2)
Revenue_playas_df = pd.DataFrame({"Total Purchase" : Revenue_playas})
```

```python
#join the data into a DF and add dollar signs
Playasdata_df = count_playasdf.join(Buyavg_playas_df)
Playasdata2_df = Playasdata_df.join(Revenue_playas_df)

#sort and select only the top 5
sorted_playas = Playasdata2_df.sort_values('Total Purchase',ascending=False)
Topspenders_df = sorted_playas.head(5)

#add dollar signs after the sort
Topspenders_df["Total Purchase"] = '$' + Topspenders_df["Total Purchase"].astype(str) 
Topspenders_df["Average Purchase Price"] = '$' + Topspenders_df["Average Purchase Price"].astype(str) 
Topspenders_df
```

```python
### Most Popular Items
#do the groupby function
pop_items = gamesdf.groupby(["Item ID", "Item Name"]) 
pop_items.head(1)
```

```python
#Purchase Count calculate all purchases based on SN of the players 
Items_count = pop_items["Item ID"].count()
Items_countdf = pd.DataFrame({"Purchase Counts" : Items_count})

# Item Price
Items_Price = pop_items["Price"].mean()
Items_Price = Items_Price.round(2)
Items_Price_df = pd.DataFrame({"Items Price" : Items_Price})

# Total Purchase Value
Item_total = pop_items['Price'].sum()
Item_total = Item_total.round(2)
Item_total_df = pd.DataFrame({"Total Purchase" : Item_total})
```

```python
#join the data into a DF 
Popularitemsdata_df = Items_countdf.join(Items_Price_df)
Popularitemsdata2_df = Popularitemsdata_df.join(Item_total_df)

#sort and select only the top 5
Popular_itemsdf = Popularitemsdata2_df.sort_values('Purchase Counts',ascending=False)

#add dollar signs after the sort
Popular_itemsdf["Total Purchase"] = '$' + Popular_itemsdf["Total Purchase"].astype(str) 
Popular_itemsdf["Items Price"] = '$' + Popular_itemsdf["Items Price"].astype(str) 
Popular_itemsdf.head(5)
```

```python
### Most Profitable Items
Profitable_itemsdf = Popularitemsdata2_df.sort_values("Total Purchase",ascending=False)
Profitable_itemsdf["Total Purchase"] = '$' + Profitable_itemsdf["Total Purchase"].astype(str) 
Profitable_itemsdf["Items Price"] = '$' + Profitable_itemsdf["Items Price"].astype(str) 
Profitable_itemsdf.head(5)
```
